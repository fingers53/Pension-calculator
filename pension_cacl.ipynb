```python
import matplotlib.pyplot as plt
import numpy as np

number_of_years_compound = 50
income = 27000
ROI = 1.07

def grow_income(current_income,year):
    if year < 8 and year%2 == 1:
        new_income = round(current_income* 1.33,2)
    
    elif 8 < year < 25 and year % 5 == 0:
        new_income = round(current_income*1.04,2)

    elif year > 25 and year%3 == 0:
        new_income = round(current_income* 0.95,2) 
    
    else:
        new_income = current_income
    
    return new_income

def salary_per_year(length_of_career,income,max_income=80000):
    total = 0
    pension=[]
    incomes=[]
    for year in range(length_of_career):
        income = grow_income(income,year) if income < max_income else income
        incomes.append(income)
        if year%5 == 1:
            print('Year: ',year,'Income',income,'Total pension',total) 
    plt.plot(incomes)   
    print('\nLIFETIME EARNINGS: ',sum(incomes))
    return incomes
```

```python
def pension(incomes,length_of_career,portion_of_income_to_pension=0.1,expected_roi=1.07):
    pension_fund = 0 
    pension_contributions = []
    for count,year in enumerate(range(length_of_career)):
        principal_added_per_year = portion_of_income_to_pension*incomes[count]
        pension_fund += principal_added_per_year*ROI
        pension_fund = round(pension_fund,2)
        pension_contributions.append(pension_fund)
    plt.plot(range(length_of_career),pension_contributions)
    print('PENSION AMOUNT AT RETIREMENT:',pension_fund)
    return pension_fund

```

```python
lifetime_earnings = salary_per_year(40,27_000,max_income=120_000)
```
